```
# Task 1: Clean pinterest df

# File location and type
# Asterisk(*) indicates reading all the content of the specified file that have .json extension
file_location = "/mnt/12acc47946a5/topics/12acc47946a5.pin/partition=0/*.json" 
file_type = "json"
# Ask Spark to infer the schema
infer_schema = "true"
# Read in JSONs from mounted S3 bucket
df_pin = spark.read.format(file_type) \
.option("inferSchema", infer_schema) \
.load(file_location)
# Display Spark dataframe to check its content
#display(df_pin)
```

```
# Task 1: Clean pinterest df

from pyspark.sql.functions import regexp_replace

cleaned_df_pin = df_pin.replace({'No description available Story format': None}, subset=['description'])
cleaned_df_pin = cleaned_df_pin.replace({'User Info Error': None}, subset=['follower_count'])
cleaned_df_pin = cleaned_df_pin.replace({'Image src error.': None}, subset=['image_src'])
cleaned_df_pin = cleaned_df_pin.replace({'User Info Error': None}, subset=['poster_name'])
cleaned_df_pin = cleaned_df_pin.replace({'N,o, ,T,a,g,s, ,A,v,a,i,l,a,b,l,e': None}, subset=['tag_list'])
cleaned_df_pin = cleaned_df_pin.replace({'No Title Data Available': None}, subset=['title'])

cleaned_df_pin = cleaned_df_pin.replace({'k': '000'}, subset=['follower_count'])
cleaned_df_pin = cleaned_df_pin.replace({'m': '000000'}, subset=['follower_count'])

cleaned_df_pin = cleaned_df_pin.withColumn("follower_count", regexp_replace("follower_count", "k", "000"))
                   

cleaned_df_pin = cleaned_df_pin.withColumn("follower_count", cleaned_df_pin["follower_count"].cast("int"))
cleaned_df_pin.dtypes


cleaned_df_pin = cleaned_df_pin.withColumn("save_location", regexp_replace("save_location", "Local save in ", ""))

cleaned_df_pin = cleaned_df_pin.withColumnRenamed("index", "ind")
cleaned_df_pin = cleaned_df_pin.select("ind", "unique_id", "title","description","follower_count","poster_name","tag_list","is_image_or_video","image_src","save_location","category")
display(cleaned_df_pin)

```

```
#Task 2: geolocation df cleaning
# File location and type
# Asterisk(*) indicates reading all the content of the specified file that have .json extension
file_location = "/mnt/12acc47946a5/topics/12acc47946a5.geo/partition=0/*.json" 
file_type = "json"
# Ask Spark to infer the schema
infer_schema = "true"
# Read in JSONs from mounted S3 bucket
geo_df = spark.read.format(file_type) \
.option("inferSchema", infer_schema) \
.load(file_location)
# Display Spark dataframe to check its content
display(geo_df)
```

```
#Task 2: geolocation df cleaning

from pyspark.sql.functions import concat, lit

geo_df_clean =  geo_df.withColumn("coordinates", concat("latitude", lit(","), "longitude"))
geo_df_clean = geo_df_clean.withColumn("timestamp", geo_df_clean["timestamp"].cast("timestamp"))
geo_df_clean.dtypes

geo_df_clean = geo_df_clean.select("ind","country","coordinates","timestamp")

display(geo_df_clean)

```

```
#Task 3: user df cleaning
# File location and type
# Asterisk(*) indicates reading all the content of the specified file that have .json extension
file_location = "/mnt/12acc47946a5/topics/12acc47946a5.user/partition=0/*.json" 
file_type = "json"
# Ask Spark to infer the schema
infer_schema = "true"
# Read in JSONs from mounted S3 bucket
user_df = spark.read.format(file_type) \
.option("inferSchema", infer_schema) \
.load(file_location)
# Display Spark dataframe to check its content
display(user_df)
```

```
#Task 3: user df cleaning
user_df_clean = user_df.withColumn("user_name", concat("first_name", lit(" "), "last_name"))
user_df_clean = user_df_clean.withColumn("date_joined", user_df_clean["date_joined"].cast("timestamp"))
user_df_clean = user_df_clean.select("ind","user_name","age","date_joined")


display(user_df_clean)
```

```
#Task 4: Most popular category at each conutry
from pyspark.sql.window import Window
from pyspark.sql.functions import max
import pyspark.sql.functions as f

#Queries
combined_df = cleaned_df_pin.join(geo_df_clean, geo_df_clean["ind"] == cleaned_df_pin["ind"], how="inner")
grouped_pin_df = combined_df.groupBy(["category","country"]).agg({"category": "count"})

# Creating a Window specification
window_spec = Window.partitionBy("country")#.orderBy("count(category)")

# Applying window functions
result_df = grouped_pin_df.withColumn("category_count", max("count(category)").over(window_spec))

#Keep only max counts
result_df = result_df.filter(f.col("category_count") == f.col('count(category)'))
result_df = result_df.select("country","category","category_count")
display(result_df)

```

```
#Task 5: Find most popular category each year
from pyspark.sql.functions import year

combined_df = combined_df.withColumn('post_year', year(combined_df['timestamp']))

grouped_df = combined_df.groupBy(["category","post_year"]).agg({"category": "count"})
result_df = grouped_df.filter(f.col("post_year") >= 2018).filter(f.col("post_year") <= 2022)
result_df.select("post_year","category","count(category)")
result_df = result_df.withColumnRenamed("count(category)", "category_count")
# result_df = result_df.groupBy(["post_year"]).agg({"category": "max"})

display(result_df)

# Creating a Window specification
window_spec = Window.partitionBy("post_year")

# Applying window functions
result_df = result_df.withColumn("cat-count", max("category_count").over(window_spec))
result_df = result_df.filter(f.col("cat-count") == f.col('category_count'))
result_df = result_df.select("post_year","category","category_count")
display(result_df)
```

```
#Task 6: for each country find the user with most followers
from pyspark.sql.functions import col, max

grouped_df = combined_df.groupBy(["country","follower_count","poster_name"]).agg({"follower_count": "max"})
display(grouped_df)
# country_most_followers_df = grouped_df.select(,max(grouped_df.follower_count))
# display(country_most_followers_df)

# a = grouped_df.groupby(grouped_df['follower_count']).agg({"follower_count": "max"})
max_follower_count = grouped_df.select(max("follower_count")).collect()
max_follower_row = grouped_df.filter(col("follower_count") == max_follower_count).show()
display(max_follower_row)
result_df = max_follower_row.select('country','follower_count')
display(result_df)
```

```
#Task 7: most popular category people post by age group
from pyspark.sql.functions import when
from pyspark.sql.functions import max

display(user_df_clean)
user_df_clean.dtypes

# age_group_a = (user_df_clean['age']>=18)
age_group_a = ((col('age'))>=18) & (col('age') <=24)
age_group_b = ((col('age'))>=25) & (col('age')<=35)
age_group_c = ((col('age'))>=36) & (col('age')<=50)
age_group_d = (user_df_clean['age']>50)

user_df_clean = user_df_clean.withColumn('age_group', when(age_group_a, '18-24').when(age_group_b,'25-35').when(age_group_c,'36-50').when(age_group_d,'50+').otherwise(None))
# user_df_clean = user_df_clean.withColumn("age_group", when(age_group_b, ).otherwise(None))
# user_df_clean = user_df_clean.withColumn("age_group", when(age_group_c, '36-50').otherwise(None))
# user_df_clean = user_df_clean.withColumn("age_group", when(age_group_d, '50+').otherwise(None))
display(user_df_clean)
age_pin_df_raw = cleaned_df_pin.join(user_df_clean, user_df_clean["ind"] == cleaned_df_pin["ind"], how="inner")


# age_pin_df = age_pin_df.groupby(['category','age_group']).agg({'age_group':'count'}).alias('age_group_total')
age_pin_df = age_pin_df_raw.groupby(['category','age_group']).agg(count('age_group')).alias('age_group_total')

age_pin_df = age_pin_df.withColumnRenamed("count(age_group)", "age_group_total")
# grouped_df = age_pin_df.groupBy(["category","age_group"]).agg({"age_group_total": "max"})


# Creating a Window specification
window_spec = Window.partitionBy("age_group")

# Applying window functions
result_df = age_pin_df.withColumn("category_count", max("age_group_total").over(window_spec))
display(result_df)
result_df = result_df.filter(col("category_count") == col("age_group_total"))

display(age_pin_df)
# display(grouped_df)
result_df = result_df.select("age_group","category","category_count")
display(result_df)
```

```
#Task 8: median follower count for each age group
from pyspark.sql.functions import expr

grouped_median = age_pin_df_raw.groupBy("age_group") \
    .agg(expr("percentile_approx(follower_count, 0.5)").alias("median_follower_count"))


display(grouped_median)
```

```
#Task 9: How many users joined each year
# display(user_df_clean)

user_year_joined_df = user_df_clean.withColumn('post_year', year(user_df_clean['date_joined']))
display(user_year_joined_df)

users_grouped = user_year_joined_df.groupby('post_year').agg({'post_year':'count'})
users_grouped = users_grouped.withColumnRenamed("count(post_year)", "number_users_joined")

display(users_grouped)
```

```
# Task 10: median follower count of users based on joining year
from pyspark.sql.functions import expr

user_year_joined_df = age_pin_df_raw.withColumn('post_year', year(user_df_clean['date_joined']))
# display(user_year_joined_df)

user_year_joined_group_df = user_year_joined_df.groupby('post_year').agg(expr("percentile_approx(follower_count, 0.5)"))
user_year_joined_group_df = user_year_joined_group_df.withColumnRenamed("percentile_approx(follower_count, 0.5)", "median_follower_count")

display(user_year_joined_group_df)
# grouped_median = age_pin_df_raw.groupBy("age_group") \
#     .agg(expr("percentile_approx(follower_count, 0.5)").alias("median_follower_count"))
```
